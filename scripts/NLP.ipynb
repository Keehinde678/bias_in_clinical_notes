```python
import pandas as pd
import numpy as np
import re
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import warnings
warnings.filterwarnings('ignore')

# Set styling
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("MIMIC-III LEXICON-BASED BIAS ANALYSIS")

# LOAD AND FILTER DATA
print("\n Loading and filtering data...")

df = pd.read_csv('MIMICIII_Cleaned_Merged.csv')
print(f"✓ Loaded {len(df):,} total records")

# Filter to top 5 ethnicity groups
top_5_ethnicities = [
    'WHITE',
    'BLACK/AFRICAN AMERICAN',
    'HISPANIC OR LATINO',
    'OTHER',
    'ASIAN'
]

df_filtered = df[df['ETHNICITY'].isin(top_5_ethnicities)].copy()
print(f"✓ Filtered to top 5 ethnicities: {len(df_filtered):,} records")
print("\nDistribution:")
for eth in top_5_ethnicities:
    count = len(df_filtered[df_filtered['ETHNICITY'] == eth])
    pct = (count / len(df_filtered)) * 100
    print(f"  {eth:30s}: {count:6,} ({pct:5.2f}%)")
```

```python
#  CONSTRUCT COMPREHENSIVE LEXICONS

print(" Constructing lexicons")

# H1: Compliance and Agency Terms
negative_compliance = [
    'non-compliant', 'noncompliant', 'non compliant',
    'refuses', 'refused', 'refusing', 'refusal',
    'uncooperative', 'un-cooperative', 'uncooperative',
    'non-adherent', 'nonadherent', 'non adherent',
    'resistant', 'resists', 'resisting',
    'difficult', 'demanding', 'combative',
    'agitated', 'aggressive', 'hostile',
    'unmotivated', 'unwilling', 'defiant'
]

positive_compliance = [
    'compliant', 'cooperative', 'adherent',
    'motivated', 'engaged', 'willing',
    'follows', 'agrees', 'accepting',
    'participates', 'collaborative'
]

# H3: Epistemic Markers (Doubt vs Neutral)
doubt_markers = [
    'claims', 'alleges', 'allegedly', 'states',
    'insists', 'supposedly', 'purportedly',
    'believes', 'feels', 'thinks'
]

neutral_markers = [
    'reports', 'describes', 'explains',
    'indicates', 'notes', 'mentions',
    'presents with', 'demonstrates'
]

# H4: Stereotype Descriptors (Warmth vs Competence)
warmth_descriptors = [
    'pleasant', 'friendly', 'kind', 'warm',
    'caring', 'nice', 'sweet', 'gentle'
]

competence_descriptors = [
    'intelligent', 'articulate', 'coherent',
    'oriented', 'appropriate', 'rational',
    'capable', 'competent', 'sharp'
]

# H2: Behavioral vs Clinical Focus
behavioral_words = [
    'behavior', 'attitude', 'personality', 'manner',
    'appearance', 'demeanor', 'disposition',
    'affect', 'mood', 'emotional', 'anxious',
    'angry', 'frustrated', 'upset', 'tearful'
]

clinical_words = [
    'diagnosis', 'symptom', 'treatment', 'procedure',
    'medication', 'lab', 'test', 'imaging',
    'vital signs', 'cardiac', 'respiratory',
    'assessment', 'exam', 'findings'
]

print(f" Negative compliance terms: {len(negative_compliance)}")
print(f" Positive compliance terms: {len(positive_compliance)}")
print(f" Doubt markers: {len(doubt_markers)}")
print(f" Neutral markers: {len(neutral_markers)}")
print(f" Warmth descriptors: {len(warmth_descriptors)}")
print(f" Competence descriptors: {len(competence_descriptors)}")
print(f" Behavioral words: {len(behavioral_words)}")
print(f" Clinical words: {len(clinical_words)}")
```

```python
# FEATURE EXTRACTION FUNCTION
print(" Extracting features from clinical notes")

def count_lexicon_matches(text, lexicon):
    """Count occurrences of lexicon terms in text (case-insensitive)"""
    if pd.isna(text):
        return 0
    text_lower = str(text).lower()
    count = 0
    for term in lexicon:
        # Use word boundaries to avoid partial matches
        pattern = r'\b' + re.escape(term.lower()) + r'\b'
        count += len(re.findall(pattern, text_lower))
    return count

def calculate_rate_per_10k(count, word_count):
    """Calculate rate per 10,000 words"""
    if word_count == 0:
        return 0
    return (count / word_count) * 10000

# Extract features
print("Extracting negative compliance terms...")
df_filtered['neg_compliance_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, negative_compliance)
)

print("Extracting positive compliance terms...")
df_filtered['pos_compliance_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, positive_compliance)
)

print("Extracting doubt markers...")
df_filtered['doubt_marker_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, doubt_markers)
)

print("Extracting neutral markers...")
df_filtered['neutral_marker_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, neutral_markers)
)

print("Extracting warmth descriptors...")
df_filtered['warmth_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, warmth_descriptors)
)

print("Extracting competence descriptors...")
df_filtered['competence_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, competence_descriptors)
)

print("Extracting behavioral words...")
df_filtered['behavioral_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, behavioral_words)
)

print("Extracting clinical words...")
df_filtered['clinical_count'] = df_filtered['TEXT'].apply(
    lambda x: count_lexicon_matches(x, clinical_words)
)

# Calculate word counts (already exists as 'word_count')
if 'word_count' not in df_filtered.columns:
    print("Calculating word counts...")
    df_filtered['word_count'] = df_filtered['TEXT'].astype(str).apply(
        lambda x: len(x.split())
    )

# Calculate rates per 10,000 words
print("Calculating rates per 10,000 words...")
df_filtered['neg_compliance_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['neg_compliance_count'], row['word_count']), 
    axis=1
)

df_filtered['pos_compliance_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['pos_compliance_count'], row['word_count']),
    axis=1
)

df_filtered['doubt_marker_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['doubt_marker_count'], row['word_count']),
    axis=1
)

df_filtered['neutral_marker_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['neutral_marker_count'], row['word_count']),
    axis=1
)

df_filtered['warmth_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['warmth_count'], row['word_count']),
    axis=1
)

df_filtered['competence_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['competence_count'], row['word_count']),
    axis=1
)

df_filtered['behavioral_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['behavioral_count'], row['word_count']),
    axis=1
)

df_filtered['clinical_rate'] = df_filtered.apply(
    lambda row: calculate_rate_per_10k(row['clinical_count'], row['word_count']),
    axis=1
)

print(" Feature extraction complete!")

# STEP 4: DESCRIPTIVE STATISTICS BY ETHNICITY
print("\n" + "="*80)
print("STEP 4: Calculating descriptive statistics by ethnicity...")
print("="*80)

# Group by ethnicity
results = []
for ethnicity in top_5_ethnicities:
    subset = df_filtered[df_filtered['ETHNICITY'] == ethnicity]
    
    results.append({
        'Ethnicity': ethnicity,
        'N': len(subset),
        'Neg_Compliance_Rate': subset['neg_compliance_rate'].mean(),
        'Pos_Compliance_Rate': subset['pos_compliance_rate'].mean(),
        'Doubt_Marker_Rate': subset['doubt_marker_rate'].mean(),
        'Neutral_Marker_Rate': subset['neutral_marker_rate'].mean(),
        'Warmth_Rate': subset['warmth_rate'].mean(),
        'Competence_Rate': subset['competence_rate'].mean(),
        'Behavioral_Rate': subset['behavioral_rate'].mean(),
        'Clinical_Rate': subset['clinical_rate'].mean(),
    })

results_df = pd.DataFrame(results)
results_df = results_df.round(2)

print("\nTable 1: Linguistic Features by Ethnicity (Rate per 10,000 words)")
print("="*80)
print(results_df.to_string(index=False))
```

```python

# STATISTICAL ANALYSIS (ANOVA)
print(" Statistical analysis (One-way ANOVA)")

# Prepare groups for ANOVA
groups = {eth: df_filtered[df_filtered['ETHNICITY'] == eth] for eth in top_5_ethnicities}

# Run ANOVA for each feature
features = [
    ('neg_compliance_rate', 'Negative Compliance'),
    ('pos_compliance_rate', 'Positive Compliance'),
    ('doubt_marker_rate', 'Doubt Markers'),
    ('neutral_marker_rate', 'Neutral Markers'),
    ('warmth_rate', 'Warmth Descriptors'),
    ('competence_rate', 'Competence Descriptors'),
    ('behavioral_rate', 'Behavioral Words'),
    ('clinical_rate', 'Clinical Words')
]

anova_results = []

for feature, feature_name in features:
    # Extract data for each group
    group_data = [groups[eth][feature].dropna() for eth in top_5_ethnicities]
    
    # Run ANOVA
    f_stat, p_value = f_oneway(*group_data)
    
    # Calculate effect size (eta-squared)
    # Total sum of squares
    all_data = df_filtered[feature].dropna()
    grand_mean = all_data.mean()
    ss_total = np.sum((all_data - grand_mean) ** 2)
    
    # Between-group sum of squares
    ss_between = sum(len(g) * (g.mean() - grand_mean) ** 2 for g in group_data)
    
    # Eta-squared
    eta_squared = ss_between / ss_total if ss_total > 0 else 0
    
    # Degrees of freedom
    df_between = len(top_5_ethnicities) - 1
    df_within = len(all_data) - len(top_5_ethnicities)
    
    anova_results.append({
        'Feature': feature_name,
        'F-statistic': f_stat,
        'df_between': df_between,
        'df_within': df_within,
        'p-value': p_value,
        'eta_squared': eta_squared,
        'Significant': 'Yes' if p_value < 0.05 else 'No'
    })

anova_df = pd.DataFrame(anova_results)
anova_df['F-statistic'] = anova_df['F-statistic'].round(3)
anova_df['p-value'] = anova_df['p-value'].round(4)
anova_df['eta_squared'] = anova_df['eta_squared'].round(6)

print("\nTable 2: ANOVA Results")
print(anova_df.to_string(index=False))

# POST-HOC TESTS (Tukey HSD)

print(" Post-hoc pairwise comparisons (Tukey HSD)...")

# Run Tukey HSD for negative compliance (primary outcome)
print("\nPost-hoc: Negative Compliance Rate")
tukey_neg = pairwise_tukeyhsd(
    endog=df_filtered['neg_compliance_rate'],
    groups=df_filtered['ETHNICITY'],
    alpha=0.05
)
print(tukey_neg)
```
